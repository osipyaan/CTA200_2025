Part 1

```python
def f(x):
    """
    Evaluate the function f(x) = x^3 - x^2 - 1.

    Parameters:
    ----------
    x : float or array
        Input value(s).

    Returns:
    -------
    float or array
        Value of the function at x.
    """
    return x**3 - x**2 - 1


def df(x):
    """
    Evaluate the derivative f'(x) = 3x^2 - 2x.

    Parameters:
    ----------
    x : float or array
        Input value(s).

    Returns:
    -------
    float or array
        Value of the derivative at x.
    """
    return 3*x**2 - 2*x
```

Part 2

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    Perform Newton's method to find a root of a function.

    Parameters:
    ----------
    f : function
        Function for which the root is sought (f(x) = 0).
    df : function
        Derivative of the function f.
    x0 : float
        Initial guess for the root.
    epsilon : float
        Tolerance for convergence (default = 1e-6).
    max_iter : int
        Maximum number of iterations (default = 30).

    Returns:
    -------
    float or None
        Approximate root if convergence is achieved, otherwise None.
    """

    x = x0

    for i in range(1, max_iter + 1):
        if df(x) == 0: #check that derivative is zero, so that function doesn't fail
            print("Derivative is zero. Iteration failed.")
            return None

        x = x - f(x) / df(x)

        if abs(f(x)) < epsilon:
            print("Found root in", i, "iterations")
            return x

    print("Iteration failed")
    return None
```
